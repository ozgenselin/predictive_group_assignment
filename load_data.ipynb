```python
import pandas as pd

# URLs for the separate files
train_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
test_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.test"

# Define the column names manually
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 
             'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 
             'hours-per-week', 'native-country', 'income']

# Download the data
train_df = pd.read_csv(train_url, names=col_names, sep=',', skipinitialspace=True)
test_df = pd.read_csv(test_url, names=col_names, sep=',', skipinitialspace=True, skiprows=1)

# Combine into a single DataFrame
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Clean the trailing periods from the test set labels
combined_df['income'] = combined_df['income'].str.replace('.', '', regex=False)

# Save to a single CSV file
combined_df.to_csv('adult_income.csv', index=False)

# Verify the final shape
print(f"Success! Saved combined dataset with {combined_df.shape[0]} rows and {combined_df.shape[1]} columns.")
```
